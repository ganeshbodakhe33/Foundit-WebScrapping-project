```python
import pandas as pd
import numpy as np
import time
import requests
import selenium
from selenium import webdriver
import warnings
warnings.filterwarnings('ignore')
from bs4 import BeautifulSoup
import urllib3
```

```python
def getUrl(key):
    key = key.replace(' ','+')
    temp = 'https://www.foundit.in/srp/results?query={}'
    url = temp.format(key)
    return url
```

```python
url = getUrl('python')
print(url)
```

```python
driver = webdriver.Firefox()
driver.get(url)
```

```python
def find_job(card):
    # Fetching job_name
    try:
        job_name = card.find('div','jobTitle').text
        job_name = job_name.replace('\n','')
    except:
        job_name = 'Not Mentioned'
        
    # Fetching company_name
    try:
        company_name = card.find('div','companyName').text
        company_name = company_name.replace('\n','')
    except:
        company_name = 'Not Mentioned'
        
    # Fetching company_location_and_details
    try:
        company_location_and_details = card.find('div','cardBody').text
    except:
        company_location_and_details = 'Not Mentioned'
        
    # Fetching skills_required
    try:
        skills_required = card.find('div','skillDetails').text
    except:
        skills_required = 'Not Mentioned'

#     # Fetching job_apply_url
#     try:
#         job_apply_url = card.find('a')
#         job_apply_url = 'https://www.foundit.in/srp/results?query=python' + job_apply_url['href']
#     except:
#         job_apply_url = 'Not Mentioned'
    
    all_job_detail = (job_name,company_name,company_location_and_details,skills_required)
    return all_job_detail
```

```python
all_jobs = []
for i in cards:
    temp = find_job(i)
    all_jobs.append(temp)
```

```python
# saving this data to a csv file
col = ['job_name','company_name','company_location_and_details','skills_required']
Foundit_jobs_data = pd.DataFrame(all_jobs,columns=col)
Foundit_jobs_data.to_csv('D:\Ganesh\PROJECTS\----Foundit WebScrapping\Foundit_jobs_data.csv',encoding='utf-8')
```
